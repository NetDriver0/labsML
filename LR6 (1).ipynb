```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

объявляю нужную функцию, указанную в задании

```python
def y (x1,x2):
    return (x1 + x2)
```

```python
c0 = []
f0 = []
for i in range (-100,100):
    for j in range (-100,100):
            f0.append(y(i,j))
            c0.append([i,j])
```

генерирую случайные значения с использованием указанной выше функции и заполняю ей значения

```python
c = np.array(c0)
f = np.array(f0)
```

```python
print(c[0, 0])
```

```python
model = keras.Sequential()
```

```python
model.add(Dense(units=1, input_shape=(2,), activation='linear'))
```

```python
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1))
```

```python
log = model.fit(c, f, epochs=5, verbose=False)
```

обучаю модель и смотрю на результаты по мере обучения

```python
plt.plot(log.history['loss'])
plt.grid(True)
plt.show()
```

тут может показаться что график линейный, состоящий из двух отрезков

```python
plt.plot(log.history['loss'],)
plt.xlim(0,5)
plt.ylim(0,0.00000000000000001)
plt.show()
```

Но тут мы можем убедится, что график экспоненциальный с высоким значением и просто имеет мало промежуточных значений

```python
print(model.predict(x=np.array([[10,-1]])))
```

```python
print(model.predict(x=np.array([[10000000,-10000000]])))
```
